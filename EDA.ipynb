```python
date_str = df2['Event dates']
#strip() = a method used to remove leading and trailing whitespace characters from the string
def update_date_range_format(date_str):
    if '-' in date_str:
        start_date, end_date_month_year = date_str.split('-')
        end_date, month_year = end_date_month_year.split('.', maxsplit=1)
        if(len(start_date) > 4):
            month, year = month_year.split('.')
            formatted_date_range = f"{start_date.strip()}{year.strip()} - {end_date.strip()}.{month_year.strip()}"
        else:
            formatted_date_range = f"{start_date.strip()}{month_year.strip()} - {end_date.strip()}.{month_year.strip()}"
        return formatted_date_range
    else:
        return date_str  # Return the original date string if it doesn't contain a dash

# Assuming df2['Event dates'] contains the date strings
# Apply the function to update the date format
df2['Event dates'].astype(str).apply(update_date_range_format).unique()
```

```python
df2['Event dates'] = df2['Event dates'].astype(str).apply(update_date_range_format)
```

```python
def extract_dates(date_str):
    if '-' in date_str:
        start_date, end_date = date_str.split(' - ')
        return pd.to_datetime(start_date, format="%d.%m.%Y"), pd.to_datetime(end_date, format="%d.%m.%Y")
    else:
        date = pd.to_datetime(date_str, format="%d.%m.%Y")
        return date, date

df2['start_date'], df2['end_date'] = zip(*df2['Event dates'].apply(extract_dates))
#* --> unpacking operator --> takes an iterable and unpacks its elements into seperate arguments (arg1, arg2)
#zip --> combine multiple iterables into tuples --> coverting rows into columns
```

#### Find the seasons for the data --> Are runners in summer slower than winter? split between two decimals
* Spring 3-5
* Summer 6-8
* Fall 9-11
* Winter 12-2

> Yes, @summer is slower

```python
race_month = df2['race_start'].dt.month
print(race_month)
```

```python
def get_season(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'
    else:
        return 'Winter'

race_season = race_month.apply(get_season)
print(race_season)
```

```python
df3.groupby(race_season)['athlete_avg_speed'].mean()
```
